## Optimization - Genetic Algorithm


```python
# Import the class
from Momentum_Backtester_GA import Momentum_Backtester

# Create Backtester Object
mbt = Momentum_Backtester(
    asset="BTCUSDT",  # Name matches the CSV filename
    start_time="2018-02-10 00:00:00",
    end_time="2025-03-31 23:00:00",
    capital=10000,
    t_cost=0.001  # 0.1% transaction cost, adjust if needed
)

# Choose optimization method
optimization_method = 'GA'  # or 'Brute Force'

# Run optimization
if optimization_method == 'Brute Force':
    print("Running Brute Force Optimization...")
    momentum_range = range(1, 2000)
    results = mbt.parameters_optimization(momentum_range)
    mbt.heatmap_visualisation(results)

elif optimization_method == 'GA':
    print("Running Genetic Algorithm Optimization...")
    results = mbt.genetic_algorithm_optimization(
        generations=100,     # how many times we let the population evolve and improve
        population_size=100, # how many "momentum candidates" we breed and test in each generation
        mutation_rate=0.2,   #  how much randomness we allow (to avoid getting stuck in bad solutions)
        momentum_bounds=(1, 10000)   # range of momentum values want to explore
    )
    mbt.ga_visualisation(results)  # <-- use GA visualisation

else:
    print("Invalid optimization method selected.")
```
